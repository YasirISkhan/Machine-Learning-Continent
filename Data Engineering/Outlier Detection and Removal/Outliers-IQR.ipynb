```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(16,5))

plt.subplot(1,2,1)
sns.distplot(df['cgpa'])

plt.subplot(1,2,2)
sns.distplot(df['placement_exam_marks'])

plt.show()
```

```python
sns.boxplot(data = df, x=df['placement_exam_marks'])
```

```python
df['placement_exam_marks'].describe()
```

```python
percentile25 = df['placement_exam_marks'].quantile(0.25)
percentile75 = df['placement_exam_marks'].quantile(0.75)
```

```python
percentile25
```

```python
percentile75
```

```python
IQR = percentile75 - percentile25
```

```python
IQR
```

```python
upper_limit = percentile75 + 1.5*IQR
lower_limit = percentile25 - 1.5*IQR
```

```python
upper_limit
```

```python
lower_limit
```

```python
# Trimming
df[df['placement_exam_marks'] > upper_limit]
```

```python
df[df['placement_exam_marks'] < lower_limit]
```

```python
# Removing outliers
new_df = df[df['placement_exam_marks'] < upper_limit]
```

```python
# Checking if we have the outliers now
new_df[new_df['placement_exam_marks'] > upper_limit]
```

```python
sns.boxplot(data=new_df, x=new_df['placement_exam_marks'])
```

```python
# Capping
new_frame = df.copy()
new_frame['placement_exam_marks'] = np.where(df['placement_exam_marks'] > upper_limit, upper_limit, np.where(df['placement_exam_marks'] < lower_limit, lower_limit, df['placement_exam_marks']))
```

```python
plt.figure(figsize=(16,5))

plt.subplot(1,2,1)
sns.distplot(df['placement_exam_marks'])

plt.subplot(1,2,2)
sns.distplot(new_frame['placement_exam_marks'])

plt.show()
```
