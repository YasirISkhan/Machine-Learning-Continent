# 1. scatterplot(Numerical-Numerical)

```python
sns.scatterplot(data=tips, x=tips['total_bill'], y=tips['tip'])
```

```python
sns.scatterplot(data=tips, x=tips['total_bill'], y=tips['tip'], hue=tips['sex'], style=tips['smoker'])
```

```python
sns.scatterplot(data=tips, x=tips['total_bill'], y=tips['tip'], hue=tips['sex'], style=tips['smoker'], size=tips['size'] )
```

# 2. Bar plot(Numerical-Categorical)

```python
titanic.head(2)
```

```python
sns.barplot(data=titanic, x=titanic['Pclass'], y=titanic['Age'])
```

```python
sns.barplot(data=titanic, x=titanic['Pclass'], y=titanic['Age'], hue=titanic['Sex'])
```

# 3. Boxplot(Numerical-Categorical)

```python
sns.boxplot(data=titanic, x=titanic['Sex'], y=titanic['Age'], hue=titanic['Survived'])
```

# 4. Distplot (Numerical-Categorical)

```python
sns.distplot( titanic[titanic['Survived']==0]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'], hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'], hist=False)
```

# 5. HeatMap (Categorical-Categorical)

```python
titanic.head(3)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```
