```python
def clean(data):
    data.drop(data.columns[[0,3,5,6,10]], inplace=True, axis=1)
    print("Shape after dropping categorical columns: ", data.shape)

    data.dropna(inplace=True)
    print("Shape afer removing null values: ", data.shape)

    data.drop_duplicates(inplace=True)
    print("Shape after dropping duplicates: ", data.shape)

    return data
```

```python
df = clean(df)
```

```python
df.head()
```

```python
X = df.drop(columns=['Fraudulent'])
Y = df['Fraudulent']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=2, test_size=0.2)

X_train.shape, X_test.shape, y_train.shape
```

```python
y_train = y_train.reset_index(drop=True)
```

```python
def perceptron(X_train, y_train):
    X_train = np.insert(X_train, 0, 1, axis=1)
    weights = np.ones(X_train.shape[1])
    lr = 0.1

    for i in range(1000):
        j = np.random.randint(0,X_train.shape[0])
        y_hat = step(np.dot(X_train[j], weights))
        weights = weights + lr * (y_train.iloc[j] - y_hat) * X_train[j]

    return weights[0], weights[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_, coef_ = perceptron(X_train, y_train)
```
