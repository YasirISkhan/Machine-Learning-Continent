```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1, noise=20, random_state=13)
```

```python
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X,y)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
from sklearn.linear_model import Ridge
R1 = Ridge(alpha=10)
R1.fit(X,y)
```

```python
print(R1.coef_)
print(R1.intercept_)
```

```python
R2 = Ridge(alpha=100)
R2.fit(X,y)
```

```python
print(R2.coef_)
print(R2.intercept_)
```

```python
plt.plot(X,y, 'b.')
plt.plot(X, reg.predict(X), color='red', label='alpha=0')
plt.plot(X, R1.predict(X), color='green', label='alpha=10')
plt.plot(X, R2.predict(X), color='blue', label='alpha=100')
plt.legend()
plt.plot()
```
