```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.tree import plot_tree
```

```python
X, y = make_classification(n_features=5, n_informative=5, n_redundant=0, n_clusters_per_class=1)
```

```python
df = pd.DataFrame(X, columns=['col1','col2','col3','col4','col5'])
df['target'] = y
print(df.shape)
df.head()
```

# Function for Feature Sampling

```python
def feature_sample(df, percent):
    cols = random.sample(df.columns.tolist()[:-1], int(percent*(df.shape[1]-1)))
    new_df = df[cols]
    new_df['target'] = df['target']
    return new_df
```

```python
df1= feature_sample(df, 0.8)
df1
```

```python
df2 = feature_sample(df, 0.8)
df2
```

```python
df3 = feature_sample(df, 0.8)
df3
```

```python
clf1 = DecisionTreeClassifier()
clf2 = DecisionTreeClassifier()
clf3 = DecisionTreeClassifier()
```

```python
clf1.fit(df1.iloc[:,0:4], df1.iloc[:,-1])
clf2.fit(df2.iloc[:,0:4], df2.iloc[:,-1])
clf3.fit(df3.iloc[:,0:4], df3.iloc[:,-1])
```

```python
plot_tree(clf1)
```

```python
plot_tree(clf2)
```

```python
plot_tree(clf3)
```
